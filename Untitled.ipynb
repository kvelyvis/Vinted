```python
import pandas as pd
import json
import numpy as np
```

```python
pd.options.display.max_rows = 4000
pd.set_option('max_colwidth', None)
```

```python
df = pd.read_json('/Users/kristijonas/PycharmProjects/Vinted/vinted/output/20230129_sneakers.json', lines=True)
```

```python
def try_join(l):
    try:
        return ','.join(map(str, l))
    except TypeError:
        return np.nan

df['size'] = [try_join(l) for l in df['size']]
```

```python
df['url'] = df['url'].apply(', '.join)
df['scrape_time'] = df['scrape_time'].apply(', '.join)
df['title'] = df.title.astype(str).str.replace('\[|\]|\'', '')
df['price'] = df.price.astype(str).str.replace('\[|\]|\'', '')
df['seller'] = df.seller.astype(str).str.replace('\[|\]|\'', '')
df['seller_items_count'] = df.seller_items_count.astype(str).str.replace('\[|\]|\'', '')
df['description'] = df.description.astype(str).str.replace('\[|\]|\'', '')
df['brand'] = df.brand.astype(str).str.replace('\[|\]|\'', '')
df['condition'] = df.condition.astype(str).str.replace('\[|\]|\'', '')
df['colour'] = df.colour.astype(str).str.replace('\[|\]|\'', '')
df['location'] = df.location.astype(str).str.replace('\[|\]|\'', '')
df['payment'] = df.payment.astype(str).str.replace('\[|\]|\'', '')
df['views'] = df.views.astype(str).str.replace('\[|\]|\'', '')
df['uploaded'] = df.uploaded.astype(str).str.replace('\[|\]|\'', '')
df['category'] = df.category.astype(str).str.replace('\[|\]|\'', '')
df['status'] = df['status'].apply(', '.join)
df['interested'] = df.interested.astype(str).str.replace('\[|\]|\'', '')
```
